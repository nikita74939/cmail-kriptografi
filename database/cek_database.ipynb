```python
import sqlite3
import pandas as pd
```

```python
# Hubungkan ke database
conn = sqlite3.connect("cmail.db")
```

```python
# Cek daftar tabel
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
display(tables)
```

ISI TABEL USERS

```python

users_df = pd.read_sql_query("SELECT * FROM users;", conn)
users_df
```

ISI TABEL MESSAGE

```python
message_df = pd.read_sql_query("SELECT * FROM messages;", conn)
message_df
```

ISI TABEL FILES

```python

files_df = pd.read_sql_query("SELECT * FROM files;", conn)
files_df
```

ISI TABEL SQLITE_SEQUENCE

```python
sq_df = pd.read_sql_query("SELECT * FROM sqlite_sequence;", conn)
sq_df
```
